```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

```python
#load the dataset from the csv file using pandas
credit_card_data = pd.read_csv('data/creditcard.csv')
```

```python
#print the first five rows of the dataframe
credit_card_data.head()
```

```python
#determine the number of rows and columns in the dataset
credit_card_data.shape
```

```python
#dataset information
credit_card_data.info()
```

```python
#check for missing values
credit_card_data.isna().sum()
```

```python
#check the distribution of the transactions, benign or fraudulent
credit_card_data['Class'].value_counts()
```

this dataset is highly imbalanced
0 --normal transaction
1 -- fraudulent

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
#under-sampling to deal with the imbalance in the dataset
legit_sample = legit.sample(n=492)
```

```python
#concatenate the dataframes
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#split the data into features and targets
X = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
#split the data into training and testing data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=2)
```

```python
#train the model
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
#evaluate the model with accuracy score
train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(train_prediction, y_train)
```

```python
print('Training accuracy: ', training_accuracy)
```

```python
test_prediction = model.predict(X_test)
test_accuracy = accuracy_score(test_prediction, y_test)
```

```python
print('Test accuracy: ', test_accuracy)
```
